```python
async def data_from_url():
    # Set up the driver
    #driver = webdriver.Chrome(executable_path='/Users/alevi/Downloads/chromedriver_mac64/chromedriver')
    # Set up the driver with headless option
#     chrome_options = Options()
#     chrome_options.add_argument('--headless')
#     chrome_options.add_argument('--disable-gpu')
#     driver = webdriver.Chrome('/Users/alevi/Downloads/chromedriver_mac64/chromedriver', options=chrome_options)
    driver = webdriver.Chrome(executable_path=r'C:\Users\aConda\Downloads\chromedriver_win32\chromedriver')
    # Define the URLs to scrape
    global product_urls 
    global titles
    global stars
    global ratings_count
    global prices
    global orders_count
    global seller_rating
    global seller_followers
    global shipping
    global likes
    global discounts

    #Navigate to the AliExpress website and get all the product item URLs
    # driver.get(base_url)
    # wait = WebDriverWait(driver, 10)
    #i=1
    for page in range(2, 6):
        url = f"https://www.aliexpress.com/category/100003109/women-clothing.html?category_redirect=1&page={page}&spm=a2g0o.home.101.1.3b8421457304jw"
        driver.get(url)
        wait = WebDriverWait(driver, 10)
        doc_height = driver.execute_script("return document.body.scrollHeight")
        win_height = driver.execute_script("return window.innerHeight")
        num_pages = int(doc_height / win_height)
        print(f'doc height=>{doc_height}\tpages =>{num_pages}')
        # scroll through the document
        for page in range(num_pages):
            driver.execute_script("window.scrollTo(0, arguments[0]);", win_height * (page+1))
            print(f'scrolling to=>{win_height * (page+1)}')
            sleep(2)

        wait.until(EC.visibility_of_element_located((By.CLASS_NAME, "manhattan--container--1lP57Ag")))

        product_items = driver.find_elements_by_class_name("manhattan--container--1lP57Ag")

        for item in product_items:
            product_urls.append(item.get_attribute("href"))

        # Loop through each product item URL and fetch the price using BeautifulSoup

        for url in product_urls:
            driver.get(url)
            soup = BeautifulSoup(driver.page_source, 'html.parser')
            asession = AsyncHTMLSession()
            r = await asession.get(url)
            await r.html.arender(timeout=30000)
            ##stars##
            star_obj = r.html.find('.overview-rating-average')
            if len(star_obj) > 0:
                star = star_obj[0].text
            else:
                star = None
            stars.append(star)
            #####
            ##rating_counts##
            rating_obj = r.html.find('.product-reviewer-reviews')
            if len(rating_obj) > 0:
                rating_counts = rating_obj[0].text.replace(' Reviews','').strip()
            else:
                rating_counts = None
            ratings_count.append(rating_counts)
            ####
            ##prices##
            price_obj = r.html.find('.product-price-value')
            if len(price_obj) > 0:
                price = price_obj[0].text.replace('US $','').strip()
            else:
                price_obj = r.html.find('.uniform-banner-box-price')
                if len(price_obj) > 0:
                    price = price_obj[0].text.replace('US $','').strip()
                else:
                    price = None
            prices.append(price)
            ####
            ##orders_count##
            orders_count_obj = r.html.find('.product-reviewer-sold')
            if len(orders_count_obj) > 0:
                orders = orders_count_obj[0].text.replace('+ Orders','').strip()
            else:
                orders = None
            orders_count.append(orders)
            ##seller_rating##
            seller_rating_obj = r.html.find('span[data-role="positive-feedback"] i')
            if len(seller_rating_obj) > 0:
                seller_rating_text = seller_rating_obj[0].text.replace('%','').strip()
            else:
                seller_rating_text = None
            seller_rating.append(seller_rating_text)
            ##seller_followers##
            seller_follows_obj = r.html.find('p[class="num-followers"] i')
            if len(seller_follows_obj) > 0:
                seller_followers_text = seller_follows_obj[0].text.strip()
            else:
                seller_followers_text = None
            seller_followers.append(seller_followers_text)
            ####
            ##shpping##
            shipping_obj = r.html.find('strong:contains("Free Shipping")')
            if len(shipping_obj) > 0:
                shipping_stat = shipping_obj[0].text
                if shipping_stat == "Free Shipping":
                    shipping_res = 1
                else:
                    shipping_res = 0
            else:
                shipping_res = 0
            shipping.append(shipping_res)
            ####
            ##likes##
            likes_obj = r.html.find('span.add-wishlist-num')
            if len(likes_obj) > 0:
                likes_count = likes_obj[0].text
            else:
                likes_count = None
            likes.append(likes_count)
            ####
            ##titles##
            title_obj = r.html.find('.product-title-text')
            if len(title_obj) > 0:
                title = title_obj[0].text
            else:
                title = None
            titles.append(title)
            ####
            ##discounts##
            discount_obj = r.html.find('span.product-price-mark')
            if len(discount_obj) > 0:
                discount = 1
            else:
                discount = 0
            discounts.append(discount)
        product_urls.clear()
    driver.quit()
```

```python
def load_csv(file_name):
    df = pd.read_csv(file_name)
    return df
data = load_csv("data.csv")
df_copy = data.copy()
df_copy.info()
print()
print("Number of nulls")
print("----------------------------------")
null_counts = df_copy.isnull().sum()
print(null_counts)
print("----------------------------------")
```

```python
df_copy.drop_duplicates(keep='first',inplace=True)

## Titels ##
df_copy.dropna(subset=['Titles'], inplace=True)

## Prices ##
#Drop nulls, fixed price string, convert string to float 
df_copy.dropna(subset=['Prices'], inplace=True)
df_copy = df_copy[df_copy['Prices'].str.startswith('$')]
df_copy['Prices'] = df_copy['Prices'].replace({'\$': '', ',': ''}, regex=True).astype(float)

## Orders_Count ##
#replaces any non-numeric characters with an empty string and fills any remaining null (NaN) values in the column with the value 0
df_copy['Orders_Count'] = df_copy['Orders_Count'].str.replace(r'[^0-9]', '', regex=True).fillna(0).astype(int)

## Likes ##
# Fill null values with 0 and then calculate it to int 
df_copy['Likes'] = df_copy['Likes'].apply(lambda x: int(float(x.replace('K', '')) * 1000) if isinstance(x, str) and 'K' in x else int(x) if pd.notnull(x) else 0)

##Convert Start and Seller_Rating to float
df_copy['Stars'] = df_copy['Stars'].astype(float)
df_copy['Seller_Rating'] = df_copy['Seller_Rating'].astype(float)

## Convert Seller_Followers, Shipping, Discounts to int
df_copy['Seller_Followers'] = df_copy['Seller_Followers'].astype(int)
df_copy['Shipping'] = df_copy['Shipping'].astype(int)
df_copy['Discounts'] = df_copy['Discounts'].astype(int)

## Saving clean data ##
df_copy.to_csv("clean_data.csv", index=False)

df_copy
num_duplicates = df_copy.duplicated().sum()
print("Number of duplicates:", num_duplicates)
```

```python
columns_to_drop = ['Titles']
df_final = df_copy.drop(columns_to_drop, axis=1)
# Step 2: Split the Data
X = df_final.drop('Stars', axis=1)  # Features
y = df_final['Stars']  # Target variable

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
# Step 3: Feature Scaling (Normalization)
scaler = MinMaxScaler()

# Fit scaler on the training data and transform both training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 4: Training the Model
model = DecisionTreeRegressor()

# Define the parameter grid for tuning
param_grid = {'max_depth': [1, 5, 10, 15, 20, 25, 30, None],
              'min_samples_split': [2, 4, 6, 8, 10]}

# Perform grid search for parameter tuning
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5)
grid_search.fit(X_train_scaled, y_train)

# Retrieve the best hyperparameters
best_max_depth = grid_search.best_params_['max_depth']
best_min_samples_split = grid_search.best_params_['min_samples_split']

# Step 5: Model Evaluation
# Fit the model with the best hyperparameters
model = DecisionTreeRegressor(max_depth=best_max_depth, min_samples_split=best_min_samples_split)
model.fit(X_train_scaled, y_train)

# Predict on the training set
y_train_pred = model.predict(X_train_scaled)
mse_train = mean_squared_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

# Predict on the test set
y_test_pred = model.predict(X_test_scaled)
mse_test = mean_squared_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)

# Output the results
print("Best max_depth:", best_max_depth)
print("Best min_samples_split:", best_min_samples_split)
print("Mean Squared Error (MSE) on training data:", mse_train)
print("R-squared (R2) on training data:", r2_train)
print("Mean Squared Error (MSE) on test data:", mse_test)
print("R-squared (R2) on test data:", r2_test)
```
